```python
# standard imports
import pandas as pd
import json
import requests
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import kstest
```

```python
self_curation_df = pd.read_csv('./self_curation_true.csv')
no_self_curation_df = pd.read_csv('./self_curation_false.csv')
query_fee_df = pd.read_csv('../query_fees/allocationCollected.csv')
```

```python
self_curation_df.drop('Unnamed: 0', axis = 1, inplace = True)
no_self_curation_df.drop('Unnamed: 0', axis = 1, inplace = True)
self_curation_df.head()
```

```python
no_self_curation_df.head()
```

```python
query_fee_df.head()
```

```python
self_unique_subgraphs = self_curation_df['subgraphDeploymentId'].unique()
no_unique_subgraphs = no_self_curation_df['subgraphDeploymentId'].unique()
```

```python
query_fee_self_df = query_fee_df[query_fee_df['subgraphDeploymentID'].isin(self_unique_subgraphs)]
query_fee_no_df = query_fee_df[query_fee_df['subgraphDeploymentID'].isin(no_unique_subgraphs)]
```

```python
query_fee_self_df['rebateFees'] = query_fee_self_df['rebateFees'].astype(float)
query_fee_no_df['rebateFees'] = query_fee_no_df['rebateFees'].astype(float)
```

```python
query_fee_self_df['query_per_epoch'] = query_fee_self_df.groupby('epoch')['rebateFees'].cumsum()
query_fee_self_df
```

```python
query_fee_no_df['query_per_epoch'] = query_fee_no_df.groupby('epoch')['rebateFees'].cumsum()
```

```python
query_fee_no_df
```

```python
kstest(query_fee_self_df['rebateFees'], query_fee_no_df['rebateFees'])
```

```python
self_query_fee_list = []
no_query_fee_list = []
for i in range(10, 491):
    if len(query_fee_self_df[query_fee_self_df['epoch'] == i]) != 0:
        self_query_fee_list.append(list(query_fee_self_df[query_fee_self_df['epoch'] == i]['query_per_epoch'])[-1])
    else:
        self_query_fee_list.append(0)
    if len(query_fee_no_df[query_fee_no_df['epoch'] == i]) != 0:
        no_query_fee_list.append(list(query_fee_no_df[query_fee_self_df['epoch'] == i]['query_per_epoch'])[-1])
    else:
        no_query_fee_list.append(0)
```
