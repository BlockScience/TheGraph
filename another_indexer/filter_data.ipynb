# Filter all event data to a single indexer

```python
import pandas as pd
```

```python
# read in all event data
event_df = pd.read_csv('all_events/combinedEvents.csv', low_memory = False)
event_df.head()
```

```python
# drop duplicate data
event_df.drop_duplicates(inplace = True)
```

```python
# look for active indexers
event_df['indexer'].value_counts().head(20)
```

```python
event_df[event_df['type'] == 'stakeSlasheds']
```

```python
single_indexer_df = event_df[event_df['indexer'] == '0xd133fd8e0607f5d82c91626140495ea0a31d0398'].sort_values('blockNumber')
single_indexer_df[single_indexer_df['type'] == 'stakeSlasheds']
```

```python
# save indiviual indexer data to dataframe
slashed_indexer_df = event_df[event_df['indexer'] == '0x76e84025978a0c16fc7ba483718b667388f2973c'].sort_values('blockNumber')
slashed_indexer_df = slashed_indexer_df.sort_values(['blockNumber', 'logIndex']) # sort event data to correct order

#filter out each event for the indexer
slashed_indexer_df_allocation_create = slashed_indexer_df[slashed_indexer_df['type'] == 'allocationCreateds']
slashed_indexer_df_stake_delegate = slashed_indexer_df[slashed_indexer_df['type'] == 'stakeDelegateds']
slashed_indexer_df_allocation_collect = slashed_indexer_df[slashed_indexer_df['type'] == 'allocationCollecteds']
slashed_indexer_df_allocation_close = slashed_indexer_df[slashed_indexer_df['type'] == 'allocationCloseds']
slashed_indexer_df_stake_lock = slashed_indexer_df[slashed_indexer_df['type'] == 'stakeDelegatedLockeds']
slashed_indexer_df_stake_withdraw = slashed_indexer_df[slashed_indexer_df['type'] == 'stakeDelegatedWithdrawns']
slashed_indexer_df_rebate_claim = slashed_indexer_df[slashed_indexer_df['type'] == 'rebateClaimeds']
slashed_indexer_df_stake_deposited = slashed_indexer_df[slashed_indexer_df['type'] == 'stakeDepositeds']
slashed_indexer_df_rewards_assigned = slashed_indexer_df[slashed_indexer_df['type'] == 'rewardsAssigneds']
slashed_indexer_df_delegate_params = slashed_indexer_df[slashed_indexer_df['type'] == 'delegationParametersUpdateds']
slashed_indexer_df_stake_slashed = slashed_indexer_df[slashed_indexer_df['type'] == 'stakeSlasheds']

# save individual events for the indexer
slashed_indexer_df_allocation_create.to_csv('slashed_indexer/SIallocationCreateds.csv', index = False, header = True)
slashed_indexer_df_stake_delegate.to_csv('slashed_indexer/SIstakeDelegateds.csv', index = False, header = True)
slashed_indexer_df_allocation_collect.to_csv('slashed_indexer/SIallocationCollecteds.csv', index = False, header = True)
slashed_indexer_df_allocation_close.to_csv('slashed_indexer/SIallocationCloseds.csv', index = False, header = True)
slashed_indexer_df_stake_lock.to_csv('slashed_indexer/SIstakeLockeds.csv', index = False, header = True)
slashed_indexer_df_stake_withdraw.to_csv('slashed_indexer/SIstakeWithdrawns.csv', index = False, header = True)
slashed_indexer_df_rebate_claim.to_csv('slashed_indexer/SIrebateClaimeds.csv', index = False, header = True)
slashed_indexer_df_stake_deposited.to_csv('slashed_indexer/SIstakeDepositeds.csv', index = False, header = True)
slashed_indexer_df_rewards_assigned.to_csv('slashed_indexer/SIrewardsAssigneds.csv', index = False, header = True)
slashed_indexer_df_delegate_params.to_csv('slashed_indexer/SIdelegationParametersUpdateds.csv', index = False, header = True)
slashed_indexer_df_stake_slashed.to_csv('slashed_indexer/SIstakeSlasheds.csv', index = False, header = True)

# save all events for the indexer
slashed_indexer_df.to_csv('slashed_indexer/slashedIndexer.csv', index = False, header = True)
```
