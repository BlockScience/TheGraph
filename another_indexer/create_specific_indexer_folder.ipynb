```python
import pandas as pd
import os
```

```python
# read in all event data
event_df = pd.read_csv('all_events/combinedEvents.csv', low_memory = False)
event_df.head()
```

```python
# drop duplicate data
event_df.drop_duplicates(inplace = True)
```

```python
# set specific indexer
indexer_id = '0xd133fd8e0607f5d82c91626140495ea0a31d0398'
os.mkdir(indexer_id)
single_indexer_df = event_df[event_df['indexer'] == indexer_id].sort_values('blockNumber')
single_indexer_df = single_indexer_df.sort_values(['blockNumber', 'logIndex'])
```

```python
#filter out each event for the indexer
single_indexer_df_allocation_create = single_indexer_df[single_indexer_df['type'] == 'allocationCreateds']
single_indexer_df_stake_delegate = single_indexer_df[single_indexer_df['type'] == 'stakeDelegateds']
single_indexer_df_allocation_collect = single_indexer_df[single_indexer_df['type'] == 'allocationCollecteds']
single_indexer_df_allocation_close = single_indexer_df[single_indexer_df['type'] == 'allocationCloseds']
single_indexer_df_stake_lock = single_indexer_df[single_indexer_df['type'] == 'stakeDelegatedLockeds']
single_indexer_df_stake_withdraw = single_indexer_df[single_indexer_df['type'] == 'stakeDelegatedWithdrawns']
single_indexer_df_rebate_claim = single_indexer_df[single_indexer_df['type'] == 'rebateClaimeds']
single_indexer_df_stake_deposited = single_indexer_df[single_indexer_df['type'] == 'stakeDepositeds']
single_indexer_df_rewards_assigned = single_indexer_df[single_indexer_df['type'] == 'rewardsAssigneds']
single_indexer_df_delegate_params = single_indexer_df[single_indexer_df['type'] == 'delegationParametersUpdateds']
single_indexer_df_stake_slashed = single_indexer_df[single_indexer_df['type'] == 'stakeSlasheds']

# save individual events for the indexer
single_indexer_df_allocation_create.to_csv(f'{indexer_id}/SIallocationCreateds.csv', index = False, header = True)
single_indexer_df_stake_delegate.to_csv(f'{indexer_id}/SIstakeDelegateds.csv', index = False, header = True)
single_indexer_df_allocation_collect.to_csv(f'{indexer_id}/SIallocationCollecteds.csv', index = False, header = True)
single_indexer_df_allocation_close.to_csv(f'{indexer_id}/SIallocationCloseds.csv', index = False, header = True)
single_indexer_df_stake_lock.to_csv(f'{indexer_id}/SIstakeLockeds.csv', index = False, header = True)
single_indexer_df_stake_withdraw.to_csv(f'{indexer_id}/SIstakeWithdrawns.csv', index = False, header = True)
single_indexer_df_rebate_claim.to_csv(f'{indexer_id}/SIrebateClaimeds.csv', index = False, header = True)
single_indexer_df_stake_deposited.to_csv(f'{indexer_id}/SIstakeDepositeds.csv', index = False, header = True)
single_indexer_df_rewards_assigned.to_csv(f'{indexer_id}/SIrewardsAssigneds.csv', index = False, header = True)
single_indexer_df_delegate_params.to_csv(f'{indexer_id}/SIdelegationParametersUpdateds.csv', index = False, header = True)
single_indexer_df_stake_slashed.to_csv(f'{indexer_id}/SIstakeSlasheds.csv', index = False, header = True)

# save all events for the indexer
single_indexer_df.to_csv(f'{indexer_id}/singleIndexer.csv', index = False, header = True)
```
