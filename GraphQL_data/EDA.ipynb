```python
single_indexer_df_allocation_create.to_csv('SIallocationCreateds.csv', index = False, header = True)
single_indexer_df_stake_delegate.to_csv('SIstakeDelegateds.csv', index = False, header = True)
single_indexer_df_allocation_collect.to_csv('SIallocationCollecteds.csv', index = False, header = True)
single_indexer_df_allocation_close.to_csv('SIallocationCloseds.csv', index = False, header = True)
single_indexer_df_stake_lock.to_csv('SIstakeLockeds.csv', index = False, header = True)
single_indexer_df_stake_withdraw.to_csv('SIstakeWithdrawns.csv', index = False, header = True)
single_indexer_df_rebate_claim.to_csv('SIrebateClaimeds.csv', index = False, header = True)
```
